```julia
# using Pkg
# Pkg.add("Plots")
using Plots, Statistics
```

```julia
n = 100
x = rand(n).*10
```

```julia
w = 3.
b = 5.
```

```julia
y = w .*x .+ b .+ randn(n)
```

```julia
scatter(x, y)
f(x) = w*x + b
plot!(f, 0, 10, lw =2)
```

```julia
sqrt_mean(x) = sum(abs2, x) / length(x)
mean_xy(x, y) = x'*y / length(x)

function fit(x, y)
    x̄ = mean(x)
    ȳ = mean(y)

    z = mean_xy(x, y)
    η = sqrt_mean(x)
    
    ŵ = (z - x̄*ȳ) / (η - abs2(x̄))
    b̂ = ȳ - x̄*ŵ

    return ŵ, b̂
end
```

```julia
ŵ, b̂ = fit(x, y)
```

```julia
scatter(x, y)
f(x) = w*x + b
plot!(f, 0, 10, lw =2)

g(x) = ŵ*x + b̂
plot!(g, 0, 10, lw =2)
```
